```python
import numpy as np
from scipy.io import loadmat
main = loadmat("values2.mat")["values2"]
sub = np.loadtxt("submatrix_5.csv", delimiter=",")
sub_matrix_index = -1
for i in range(main.shape[0]):
    if np.array_equal(main[i], sub):
        sub_matrix_index = i
        break
if sub_matrix_index == -1:
    print("Submatrix NOT found")
    exit()
print("Submatrix_index:", sub_matrix_index)
after = sub + (0.10 * sub)
def euclidean_distances(target, main, skip_index=None):
    distances = []
    for i in range(main.shape[0]):
        if i == skip_index:
            distances.append(np.inf)   
        else:
            distances.append(np.linalg.norm(main[i] - target))
    return np.array(distances)
distances_after = euclidean_distances(after, main, skip_index=sub_matrix_index)
after_index = int(np.argmin(distances_after))
print("AFTER 10% HIGHER DEVIANT closest index:", after_index)
print("Euclidean distance to closest submatrix:", distances_after[after_index])
```
